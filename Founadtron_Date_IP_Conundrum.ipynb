```python
import pandas as pd
%matplotlib inline
import seaborn as sb
```

```python
df = pd.read_csv("ip_date_sample.csv", delimiter="|")
df.head()
```

```python
counts = df.groupby(['ip', 'timestamp']).size().groupby(level=0).max()

counts
```

```python
foo = df.groupby(['ip', 'timestamp']).size()
foo_df = pd.DataFrame({'ip':foo.index, 'max':foo.values})
foo_df
```

```python
counts_df = pd.DataFrame({'ip':counts.index, 'max':counts.values})

counts_df
```

```python
sb.set()
sb.lmplot(x="ip", y="max", data=counts_df, fit_reg=False, size = 7)
```
